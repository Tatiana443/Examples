###1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома Если реновации дома не было, то в новом признаке поставьте 0.### 

```python
df = pd.read_csv('/kc-house-data (2).csv', encoding='1251')
df.head()
```

```python
df.shape
```

```python
df['delta_renovated'] = 0
df.head()
```

```python
yr_delta = []

for i, row in df.iterrows():
    if row['yr_renovated']:
        yr_delta.append(row['yr_renovated']-row['yr_built'])
    else:
        yr_delta.append(0)
        
len(yr_delta)
```

```python
df['delta_renovated'] = yr_delta

df[['yr_built', 'yr_renovated', 'delta_renovated']].head()
```

```python
df[df['yr_renovated'] > 0][['yr_built', 'yr_renovated', 'delta_renovated']].head()
```

###1.3 Создайте признаки года продажи, месяца продажи

```python
df['yr_sale'] = df['date'].apply(lambda x: int(x[:4]))
df['mnth_sale'] = df['date'].apply(lambda x: int(x[4:6]))

df[['date', 'yr_sale', 'mnth_sale']].head()
```

###1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df.head()
```
